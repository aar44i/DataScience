# работа с имеющимися постами

```python
df_old = df_old.loc[:, 'ID':'Створ']
df_old = df_old.drop(['От устья, км'], axis=1)
df_old.columns = ['ID', 'River', 'Station']
```

```python
df_new = df_new.loc[:, 'ID':'Створ']
df_new = df_new.drop(['От устья, км'], axis=1)
df_new.columns = ['ID', 'River', 'Station']
```

```python
# result = pd.concat(frames, keys=['x', 'y', 'z'])
gauges = pd.concat([df_new, df_old],  axis=0, sort=False)
gauges["Station"] = gauges["Station"].str.lower()
gauges["River"] = gauges["River"].str.lower()
gauges.dropna(inplace=True)
gauges
```

```python
MAX_filter_24_12 = MAXDate.loc[~MAXDate['ID'].isin(gauges['ID'])]
MAX_filter_24_12
```
