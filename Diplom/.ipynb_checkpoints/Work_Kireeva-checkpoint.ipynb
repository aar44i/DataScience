```python
import pandas as pd
```

```python
cols_to_use = ['ID','River','Station','F','Lat', 'Long', 'Zero Point','Year','Qmax','Date']
MAXDate = pd.read_csv('MAXDate_2018.xlsx - Main.csv', usecols= cols_to_use)
MAXDate.head()
```

```python
MAXDate.info()
```

```python
MAXDate.dropna(inplace=True)
MAXDate
```

```python
pd.DataFrame(MAXDate.ID.isna()).sum()
```

```python
MAXDate.info()
```

```python
MAXDate['ID'] = MAXDate.ID.astype('int64')
```

```python
MAXDate
```

```python
ERC_data = pd.read_excel('ERC_data_Russia (1).xlsx')
ERC_data.head()
```

```python
def get_number(number): # функция для нахождения имени (именем считается все что идет после Mr. Miss. Don. Master. и т.д.)
    return number.split("_")[1].split("_")[0]

ERC_data['number'] = ERC_data['filename'].apply(get_number)
ERC_data
```

```python
ERC_data.info()
```

```python
stations = pd.to_numeric(ERC_data.number)
stations = stations.drop_duplicates()
stations
# df.loc[df['column_name'].isin(some_values)]
```

```python
df_filter = MAXDate.loc[MAXDate['ID'].isin(stations)]
df_filter
```

```python
df_filter.to_excel('MAXDate_filtered.xlsx')
```
